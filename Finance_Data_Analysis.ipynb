```python
from pandas_datareader import data,wb
import pandas as pd
import numpy as np
import datetime
%matplotlib inline
```

```python
start = datetime.datetime(2008,1,1)
end = datetime.datetime(2018,1,1)
BAC = data.DataReader('BAC','stooq',start,end)
C = data.DataReader('GS','stooq',start,end)
GS = data.DataReader('GS','stooq',start,end)
JPM = data.DataReader('JPM','stooq',start,end)

MS = data.DataReader('MS','stooq',start,end)
WFC = data.DataReader('WFC','stooq',start,end)
```

```python
tickers = ['BAC','C','GS','JPM','MS','WFC']
```

```python
type(MS)
```

```python
bank_stocks = pd.concat([BAC,C,GS,JPM,MS,WFC], axis=1,keys=tickers)
bank_stocks
```

```python
bank_stocks.columns.names=['Bank Ticker','Stock Info']
```

```python
bank_stocks
```

```python
# Find max Close price for each bank stock
bank_stocks.xs(key='Close',axis=1,level='Stock Info').max()
```

```python
returns = pd.DataFrame()
for tick in tickers:
  returns[tick+' Return'] = bank_stocks[tick]['Close'].pct_change()
```

```python
returns.head()
```

```python
import seaborn as sns
sns.pairplot(returns[1::])
```

```python
# find the worst single day returns for each bank
returns.idxmin()
```

```python
# find the best single day returns for each bank
returns.idxmax()
```

```python
returns.loc['2015-01-01':'2015-12-31'].std()
```

```python
sns.distplot(returns.loc['2015-01-01':'2015-12-31']['MS Return'], color='green',bins=50)
```

```python
sns.distplot(returns.loc['2010-01-01':'2010-12-31']['C Return'], color='red',bins=50)
```
